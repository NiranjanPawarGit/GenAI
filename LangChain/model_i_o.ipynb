```python
from langchain_community.llms import Ollama
from langchain.output_parsers import PydanticOutputParser
from langchain.prompts import PromptTemplate
from pydantic import BaseModel, Field, field_validator
import json

# 1️⃣ Define Joke schema
class Joke(BaseModel):
    setup: str = Field(..., description="Question to set up a joke")
    punchline: str = Field(..., description="Answer that resolves the joke")
    
    @field_validator("setup")
    def ends_with_question_mark(cls, value):
        if not value.endswith("?"):
            raise ValueError("Setup must end with a question mark.")
        return value

# 2️⃣ Create parser
parser = PydanticOutputParser(pydantic_object=Joke)

# 3️⃣ Force Ollama to follow example format
prompt = PromptTemplate(
    template=(
        "You are a funny AI. Respond ONLY with valid JSON that matches this format:\n"
        '{{"setup": "Why did the chicken cross the road?", "punchline": "To get to the other side."}}\n'
        "Do not include any other text.\n"
        "Now make a new joke for the query: {query}"
    ),
    input_variables=["query"],
)

# 4️⃣ Model
model = Ollama(model="llama2", temperature=0.0)

# 5️⃣ Run
query = "Tell me a joke."
formatted_prompt = prompt.format(query=query)
raw_output = model.invoke(formatted_prompt)

# 6️⃣ Parse JSON
try:
    parsed_data = json.loads(raw_output)
    joke = Joke(**parsed_data)
except Exception as e:
    print("Parsing failed:", e)
    joke = None

print("\nRaw LLM Output:\n", raw_output)
print("\nParsed Joke Object:\n", joke)
```

```python
from langchain.output_parsers.json import SimpleJsonOutputParser

json_prompt = PromptTemplate.from_template(
    "Return a JSON object with `birthdate` and\
        `birthplace` key that answers the following\
            question: {question}"
)

json_parser = SimpleJsonOutputParser()

json_chain = json_prompt | model | json_parser

result_list = list(json_chain.stream({"question":\
    "When and where was Elon Mush born?" }))

print(result_list)
```

```python
from langchain.output_parsers import CommaSeparatedListOutputParser
from langchain.prompts import PromptTemplate
from langchain_community.llms import Ollama

output_parser = CommaSeparatedListOutputParser()

format_instructions = output_parser.get_format_instructions()

prompt = PromptTemplate(
    template="List five {subject}.\n{format_instructions}",
    input_variables=["subject"],
    partial_variables={"format_instructions": format_instructions}
)

query="English Premier League Teams"

output = model(prompt.format(subject=query))

parsed_result = output_parser.parse(output)

print(parsed_result)
```

```python
from langchain.prompts import PromptTemplate
from langchain.output_parsers import DatetimeOutputParser, OutputFixingParser
from langchain.chains import LLMChain
from langchain_community.llms import Ollama

base_parser = DatetimeOutputParser()
parser = OutputFixingParser.from_llm(parser=base_parser, llm=Ollama(model="llama2", temperature=0))

template = """
{format_instructions}

Question: {question}

Only respond with the datetime string in the exact required format.
"""

prompt = PromptTemplate(
    template=template,
    input_variables=["question"],
    partial_variables={"format_instructions": base_parser.get_format_instructions()},
)

chain = LLMChain(prompt=prompt, llm=Ollama(model="llama2", temperature=0))

query = "When did Neil Armstrong land on the moon in terms of GMT?"
raw_output = chain.run(query)

parsed_result = parser.parse(raw_output)

print("Raw Output:", raw_output)
print("Parsed Datetime:", parsed_result)
```
